### Question 1

```python
# Read the data into a data frame.
pitching = pd.read_csv('pitching.csv')
pitching
```

```python
# Display all pitchers in 2015 under each team sorted by their last names.
pitchers2015 = pitching.loc[pitching['yearID'] == 2015]
pitchers2015.sort_values(by = 'nameLast')
```

```python
# Display all pitchers in 2015 sorted by their ERAs. This allows you to identify the best and worst pitchers
pitchers2015.sort_values(by = 'ERA')
```

```python
# Display all pitchers in 2015 under each team sorted by their ERAs.
pitchers2015.groupby(['teamID', 'playerID']).max().sort_values(by = ['teamID', 'ERA'])
```

```python
# Display all pitchers in 2015 under each team sorted by their strike outs (SO).
pitchers2015.groupby(['teamID', 'playerID']).max().sort_values(by = ['teamID', 'SO'], ascending = [True, False])
```

```python
# Calculate the average ERA for each team, list the teams by the average ERA in a descending order.
pitchers2015.groupby('teamID').mean().sort_values(by = 'ERA', ascending = False).get('ERA')
```

```python
# Calculate the total number of home runs allowed (HRs) for each team, and list the team by the count in an ascending order.
pitchers2015.groupby('teamID').sum().sort_values(by = 'HR', ascending = False).get('HR')
```

```python
# Calculate each team’s average ERA from 2011 to 2015, and list the result as a table, 
# where each column is a year, each row is a team.
after2011pitchers = pitching.loc[pitching['yearID'] >= 2011]
after2011pitchers.groupby(['yearID','teamID']).mean().get('ERA').unstack(level = 0)
```

```python
# Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015, 
# and list the result as a table, where each column is a league and each row is a year.
after2011pitchers.groupby(['yearID', 'lgID']).mean().get('ERA').unstack(level = -1)
```

```python
# List the pitchers who had the most wins in each of the years from 2011 to 2015.
allMostWinsPitchers = pd.DataFrame()
for year in range(2011, 2016):
    theYearPitchers = pitching.loc[pitching['yearID'] == year]
    new = theYearPitchers.nlargest(1, 'W')
    allMostWinsPitchers = pd.concat([allMostWinsPitchers, new])
allMostWinsPitchers
```

#### task 1: In the year of 2011-2015, Is a team’s win-loss record related to its payrolls?

```python
# Compute the total number of wins for each of the teams in MLB over 2011-2015, sort them in a descending order.
after2011teams = teams.loc[teams['yearID'] >= 2011]
winCount = after2011teams.groupby('teamID').sum().sort_values(by = 'W', ascending = False)
winCount.get('W')
```

```python
# Compute the average payroll per year for all teams over 2011-2015, sort them in a descending order.
after2011players = players.loc[players['yearID'] >= 2011]
salary = after2011players.groupby('teamID').sum().sort_values(by = 'salary', ascending = False)
salary.get('salary')
```

```python
# Create a visualization of your choice which will all allow you to show whether a team’s winning 
# record is related to its payroll. So is a team’s winning record related to its payroll?
data = pd.merge(winCount, salary, on='teamID')
plt.scatter(data.get('W'), data.get('salary'))
```

Yes, a team's winning record is related to its payroll. The more it pay, the higher winning record it get.

#### task 2: In the year of 2011-2015, Is a player’s batting performance related to his team’s win-loss record?

```python
# Compute the Batting Averages for each of the MLB teams over 2011-2015, sort them in a descending order.
# The Batting Average is defined as Hits/At Bats.
# The average is calculated from all players in each team.
teamSum = after2011teams.groupby('teamID').sum()
battingAvg = teamSum.get('H')/teamSum.get('AB')
battingAvg.sort_values(ascending = False)
```

```python
# Create a visualization of your choice which will allow you to decide whether a team’s batting average
# is related to its win-loss record. So is a team’s winning record related to its batting average?
winRate = teamSum.get('W')/(teamSum.get('W')+teamSum.get('L'))
plt.scatter(winRate, battingAvg)
```

Yes, if a team has more batting average, its win rate is higher.

#### task 3: In the year of 2011-2015, is a team’s win-loss record related to its pitching performance?

```python
# Display the average ERA (Earned Run Average) for each of the MLB teams in 2011-2015,
# sort them in a descending order. A lower ERA indicates a better pitching performance
teamMean = after2011teams.groupby('teamID').mean()
teamERA = teamMean.get('ERA')
teamERA.sort_values(ascending = False)
```

```python
# Create a visualization of your choice which will allow you to decide if a team’s win-loss record is related
# to its pitching performance. So is a team’s winning record related to its pitching performance?
plt.scatter(winRate, teamERA)
```

Yes, if a team's ERA is lower, it will have more win rate.
